# Multivariable Linear Regression

```python
import tensorflow as tf
```

## Bruteforce way implementing multivariable linear regression

```python
x1_data = [73, 93, 89, 96, 73]
x2_data = [80, 88, 91, 98, 66]
x3_data = [75, 93, 90, 100, 70]
y_data = [152, 185, 180, 196, 142]

x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)

Y = tf.placeholder(tf.float32)

w1 = tf.Variable(tf.random_normal([1]), name = 'weight1')
w2 = tf.Variable(tf.random_normal([1]), name = 'weight2')
w3 = tf.Variable(tf.random_normal([1]), name = 'weight3')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

hypothesis = x1 * w1 + x2 * w2 + x3 * w3 + b # Tensor
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y)) # Tensor
optimizer = tf.train.GradientDescentOptimizer(learning_rate= 1e-5) # Optimizer
train = optimizer.minimize(cost) # Operation

sess = tf.Session()
sess.run(tf.global_variables_initializer())
"""
Role of each node
    cost : to print cost of predicted y
    hypothesis : to print value of each dimension of hypothesis
    train : main operation to train model
"""
for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train], # Tensor, Tensor, Operation return Array, Array, None
                                   feed_dict = {x1 : x1_data, x2:x2_data, x3: x3_data, Y:y_data})
                                  
    if step % 100 == 0:
        print(step, "Cost :", cost_val, "\nPrediction:", hy_val)
```

## Clever way implementing Multivariable Regression

```python
x_data = [[73, 80, 75], [93, 88, 93], [89, 91, 90], [96, 98, 100], [73, 66, 70]]
y_data = [[152], [185], [180], [196], [142]]

X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3, 1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

hypothesis = tf.matmul(X, W) + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y)) # Tensor
optimizer = tf.train.GradientDescentOptimizer(learning_rate= 1e-5) # Optimizer
train = optimizer.minimize(cost) # Operation

hypothesis = tf.matmul(X, W) + b

sess = tf.Session()
sess.run(tf.global_variables_initializer())
"""
Role of each node
    cost : to print cost of predicted y
    hypothesis : to print value of each dimension of hypothesis
    train : main operation to train model
"""
for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train], # Tensor, Tensor, Operation return Array, Array, None
                                   feed_dict = {X: x_data, Y:y_data})
                                  
    if step % 100 == 0:
        print(step, "Cost :", cost_val, "\nPrediction:", hy_val)
```

## Building Import Graph

### String Input Producer - TextLineReader - Decoder - Batch

```python
filename = tf.train.string_input_producer(['test_score.csv'], name='filename', shuffle=False)

reader = tf.TextLineReader()
key, value = reader.read(filename)

record = [[0.], [0.], [0.], [0.]]
xy = tf.decode_csv(value, record_defaults=record)
```

```python
train_x_batch, train_y_batch = tf.train.batch([xy[0:-1], xy[-1:]], batch_size=10)

X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3, 1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

hypothesis = tf.matmul(X, W) + b

cost = tf.reduce_mean(tf.square(hypothesis - Y)) # Tensor
optimizer = tf.train.GradientDescentOptimizer(learning_rate= 1e-5) # Optimizer
train = optimizer.minimize(cost) # Operation

sess = tf.Session()
sess.run(tf.global_variables_initializer())

coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=sess, coord=coord)

for step in range(2001):
    x_batch, y_batch = sess.run([train_x_batch, train_y_batch])
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                   feed_dict = {X: x_batch, Y:y_batch})
    if step % 100 == 0:
        print(step, "Cost :", cost_val, "\nPrediction:", hy_val)
    
coord.request_stop()
coord.join(threads)
```
